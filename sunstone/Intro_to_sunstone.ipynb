#### Particular polarization states


```python
un = sun.unpol(1)
print(un.d)

lin = sun.lhp(1)
print(lin.d)
lin.plot_poincare()
```

```python
lin = sun.lvp(1)
print(lin.d)
lin.plot_poincare()
```

```python
vec2 = sun.lvp(1.5)+sun.lhp(1.)
print(vec2.d)
print(vec2.dp())
print(vec2.is_pure())
print(vec2.chi())
print(vec2.psi())
print(vec2.delta())
vec2.plot_poincare()
```

```python
M = sun.Generator(2.,5.,3.)
print(M.d)

print("Vector de salida")
print(M(vec2).d)
sun.plot_poincare(vec2)
```
